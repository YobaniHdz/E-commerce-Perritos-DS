```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/ventas.csv")
df["fecha"] = pd.to_datetime(df["fecha"])

df.head()
```

```python
ingresos_totales = df["ingresos"].sum()
print("Ingresos totales:", ingresos_totales)
```

```python
ventas_canal = df.groupby("canal")["ingresos"].sum()

ventas_canal.plot(kind="bar")
plt.title("Ingresos por canal")
plt.show()
```

```python
top_productos = df.groupby("producto")["ingresos"].sum().sort_values(ascending=False)

top_productos.plot(kind="bar")
plt.title("Top productos")
plt.show()
```

```python
ventas_dia = df.groupby("fecha")["ingresos"].sum()

ventas_dia.plot()
plt.title("Tendencia de ventas")
plt.show()
```
